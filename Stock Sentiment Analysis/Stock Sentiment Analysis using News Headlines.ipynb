## Reading Data into Dataframe

```python
import pandas as pd
```

```python
df=pd.read_csv(r'stock_data.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

### Label 1: Stock price will increase with respect to particular headlines
### Label 0: Stock price will either Constant or Decrease

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
data=train.iloc[:,2:27]
```

```python
data.columns
```

### Removing punctuations
### Removing all other except the A-z or a-z letters

```python
data.replace("[^a-zA-Z]"," ", inplace=True)
```

```python
data.head()
```

### Renaming column names for ease of access

```python
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
```

```python
data.columns= new_Index
#print(data.index)
#print(len(data))
#print(data['0'])
data.head(5)
```

### Convertng headlines to lower case
### We only apply function on a column only

```python
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
## print entire data of 2nd row
data.iloc[1,0:25]
```

### Create a list of all headlines for row 2,as 25 headlines in row 2

```python
' '.join([str(x) for x in data.iloc[1,0:25]])
```

### Create a list headlines to insert all the headlines of all rows

```python

headlines=[]
for row in range(0,len(data)):
    headlines.append(' '.join([str(i) for i in data.iloc[row,0:25]]))
    
headlines[0:3]
```

### Implement Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implement BAG OF WORDS
cv=CountVectorizer(ngram_range=(2,2))
traindata_x=cv.fit_transform(headlines)
```

### Implement RandomForest Classifier 

```python
# almost takes 2 mins
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindata_x,train['Label'])
```

### Predict for the Test Dataset

```python

test_transform= []
for row in range(0,len(test)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
```

```python
test_data = cv.transform(test_transform)
```

```python
randomclassifier.fit(traindata_x,train['Label'])
```

```python
predictions = randomclassifier.predict(test_data)
```

### Plot_confusion_matrix

```python
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
def plot_confusion_matrix(cm, title='Confusion matrix', cmap=plt.cm.Blues, labels=["positive", "negative"]):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(labels))
    plt.xticks(tick_marks, labels, rotation=45)
    plt.yticks(tick_marks, labels)
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
```

```python
plt.figure()
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(test['Label'],predictions)
print(cm)
plot_confusion_matrix(cm)    
plt.show()
```

### Checking Performance

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
score=accuracy_score(test['Label'],predictions)
print("Score :",score)
report=classification_report(test['Label'],predictions)
print(report)
```

### Implement MultinomialNB

```python
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
nb.fit(traindata_x,train['Label'])

predictions = nb.predict(test_data)
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print("Score :",score)
report=classification_report(test['Label'],predictions)
print(report)
```
